```r
library(ComplexHeatmap)
library(circlize)
library(colorspace)
library(GetoptLong)
```

```r
# load files
k4towceratio <- read.csv("./plotting/K4toWCEratio_filtered_all.csv")
k4towcereadcount <- read.csv("./plotting/K4toWCE_readcount.csv")
k4ratio <- read.csv("./plotting/K4ratio_filtered_all.csv")
wceratio <- read.csv("./plotting/WCEratio_filtered_all.csv")

# specify index column
row.names(k4towceratio) <- k4towceratio$X
row.names(k4towcereadcount) <- k4towcereadcount$start
row.names(k4ratio) <- k4ratio$X
row.names(wceratio) <- wceratio$X

# only need the following columns
k4towceratio <- k4towceratio[,2:11]
k4towcereadcount <- k4towcereadcount[,2:11]
k4ratio <- k4ratio[,2:11]
wceratio <- wceratio[,2:11]

# convert to matrix
k4towceratio_matrix <- data.matrix(k4towceratio)
k4towcereadcount_matrix <- data.matrix(k4towcereadcount)
k4ratio_matrix <- data.matrix(k4ratio)
wceratio_matrix <- data.matrix(wceratio)
```

```r
rowsep <- c(wceratio$region)
# rowsep[rowsep == 1] = '1_Very Early'
# rowsep[rowsep<=3 & rowsep >=2] = '23_Early'
# rowsep[rowsep>=4 & rowsep <=5] = '45_Middle'
# rowsep[rowsep >= 6 & rowsep <= 7] = '67_Late'
# rowsep[rowsep ==8 ] = '8_Asynchronized'
```

```r
# colors <- colorRampPalette(c("blue","navy",'black',"yellow3","yellow"))(n=255)
colors <- colorRamp2(seq(0,2, length = 5), c("blue","navy",'black',"yellow3","yellow"))

hm <- Heatmap(wceratio[1:9],
        name = "k4 ratio",
        split=rowsep,
        col = colors,
        cluster_columns = FALSE,
#         cluster_rows = FALSE,
        show_row_names = FALSE,
        width = unit(10,'cm'),
        clustering_distance_rows = "maximum",
        column_names_gp = gpar(fontsize = 10), 
        row_title_rot = 0,
        row_title_gp = gpar(fontsize=10)
       )
# set_component_height(hm, 1,unit(10,'cm'))
hm
```
